```python
%matplotlib inline 

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
x1 = np.arange(0,2880)
y1 = np.cos(x1 * np.pi / 180.0)
x2 = x1 + 0.5
y2 = y1 + 0.5
x3 = x2 + 0.5
y3 = y2 + 0.5
x4 = x3 + 0.5
y4 = y3 + 0.5
# x2 = x1
# y2 = np.tan(x2 * np.pi / 180.0)
plt.plot(x1,y1,lw=4)
plt.plot(x2,y2,'y--')
plt.plot(x3,y3,lw=2)
plt.plot(x4,y4,'y--', lw=1)
plt.xlabel("x-axis") 
plt.ylabel("y-axis") 
plt.show()
```
